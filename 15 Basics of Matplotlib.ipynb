```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lets create a basic plot to start working with 
height = [150, 160, 165, 185]
weight = [70, 80, 90, 100]
```

```python
plt.plot(height,weight)
```

```python
# add title , labels and legends 
plt.plot(height,weight)
plt.title("Relationship between height and weight")
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
calories_burnt = [65, 75, 95, 99]
plt.plot(calories_burnt)
plt.plot(weight)
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')
plt.xticks(ticks=[0,1,2,3], labels=['A','B','C','D'])
```

```python
# Size of figures, colors, markers, line style 

#figures size in inches 
plt.figure(figsize=(15,5))

plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')
plt.xticks(ticks=[0,1,2,3], labels=['A','B','C','D'])
plt.show()
```

```python
plt.plot(calories_burnt, 'go--')
plt.plot(weight, 'yo--')
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')
plt.xticks(ticks=[0,1,2,3], labels=['A','B','C','D'])
plt.show()
```

```python
plt.plot(calories_burnt, 'g')
plt.plot(weight, 'y')
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')
plt.xticks(ticks=[0,1,2,3], labels=['A','B','C','D'])
plt.show()
```

```python
#create 2 plots 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
```

```python
#plot on 0 row and 0 column 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[0,0].plot(calories_burnt, 'go')
```

```python
#plot on 0 row and 1 column 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[0,1].plot(weight, 'b--')
```

```python
#plot on 1 row and 0 column 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[1,0].plot(calories_burnt, 'ro--')
```

```python
#plot on 1 row and 0 column 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[1,1].plot(weight, 'bo--')
```

```python
#set title for subplots 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[0,0].set_title("Calories Burnt")
ax[0,1].set_title("Weight")

ax[0,0].plot(calories_burnt, 'go')
ax[0,1].plot(weight, 'b--')
ax[1,0].plot(calories_burnt, 'ro--')
ax[1,1].plot(weight, 'bo--')
```

```python
#set title for subplots 
fig, ax = plt.subplots(nrows=3, ncols=3, figsize = (9,9))
ax[0,0].set_title("Calories Burnt")
ax[0,1].set_title("Weight")

ax[0,0].plot(calories_burnt, 'go')
ax[0,1].plot(weight, 'b--')
ax[1,0].plot(calories_burnt, 'ro--')
ax[1,1].plot(weight, 'bo--')

ax[0,0].set_xticks(ticks = [0,1,2,3]);
ax[0,1].set_xticks(ticks = [0,1,2,3]);
```

```python
#set title for subplots 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (9,9), sharey=True)
ax[0,0].set_title("Calories Burnt")
ax[0,1].set_title("Weight")

ax[0,0].plot(calories_burnt, 'go')
ax[0,1].plot(weight, 'b--')
ax[1,0].plot(calories_burnt, 'ro--')
ax[1,1].plot(weight, 'bo--')

ax[0,0].set_xticklabels(labels = ['A','B','C','D'])
ax[0,1].set_xticklabels(labels = ['A','B','C','D'])
```
